# Single isotropic antenna

```python
# Reference link on colormap: https://matplotlib.org/tutorials/colors/colormaps.html
import matplotlib.pyplot as plt
import numpy as np
from numpy import log10, pi, sqrt

def channel_gain(x, y, f):
    return 10*log10((3*10**8/f)**2/(4*pi)**2) - 20*log10(sqrt(x**2+y**2))

x = np.linspace(-100, 100, 101)
y = np.linspace(-100, 100, 101)
x, y = np.meshgrid(x,y)
z = channel_gain(x, y, 50*10**6) # f = 10 MHz
color = 'YlOrRd' 
plt.figure(figsize=(6, 5))
plt.pcolormesh(x, y, z, cmap=plt.get_cmap(color), vmin=-60, vmax=-10)
plt.colorbar()
plt.contour(x,y,z, cmap=plt.get_cmap(color))
plt.xlabel('Distance [m]')
plt.ylabel('Distance [m]')
```

# Four antennas

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import log10, pi, sqrt

def channel_gain(x, y, f):
    return 10*log10((3*10**8/f)**2/(4*pi)**2) - 20*log10(sqrt(x**2+y**2))

x = np.linspace(-100, 100, 200)
y = np.linspace(-100, 100, 200)
x, y = np.meshgrid(x,y)
a = 5*10**(-2)
z = channel_gain(x, y-a/2, 50*10**6)  \
    + channel_gain(x, y+a/2, 50*10**6) \
    + channel_gain(x, y-3*a/2, 50*10**6) \
    + channel_gain(x, y+3*a/2, 50*10**6) # f = 50 MHz
color = 'YlOrRd' 
plt.figure(figsize=(6, 5))
plt.pcolormesh(x, y, z, cmap=plt.get_cmap(color), vmin=-60, vmax=-10)
plt.colorbar()
plt.contour(x,y,z, cmap=plt.get_cmap(color))
plt.xlabel('Distance [m]')
plt.ylabel('Distance [m]')
```
